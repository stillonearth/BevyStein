```python
import env
from matplotlib import pyplot as plt

%matplotlib inline 
```

```python
NUM_AGENTS = 16
ACTION_SIZE = 8
STATE_SIZE = (84, 84, 1)
```

```python
bevy_rl_shooter = env.Environment((STATE_SIZE[0], STATE_SIZE[1]), NUM_AGENTS)
```

This environment has 2 state representations: 
  - json — [agent_name, agent_name, position, rotation]
  - img — agent camera pixel

```python
bevy_rl_shooter.state()['actors'][:5]
```

Here's images of what the agent sees. bevy_rl's API returns RGB image and python wrapper downsamples it to 64x64x1.

```python
for i in range(2):
    plt.imshow(bevy_rl_shooter.visual_observations()[i], interpolation='nearest')
    plt.show()
```

Let's take a step in the environment. We can see that the agent moved forward and rotated.

```python
new_state = bevy_rl_shooter.step([
    1, 2, 3, 4, 
    1, 2, 0, 1,
    5, 1, 2, 0,
    1, 2, 0, 0])
```
